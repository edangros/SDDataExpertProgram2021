```python
#이런 프로그램을 생각해봅시다.
#1. 플레이어의 입력을 받아 틀린 입력을 받으면 에러메세지를 출력시킴
#2. 플레이어의 이름, 나이를 수집받음
#3. 최종 결과는 아래와 같은 형태로 저장
#<이름>
#나이 : n세

#루프문
while True:
    try:
        p_name = input("이름을 입력하세요 :")
        #이름이 숫자일 리 없잖아. 이름을 똑바로 입력했으면 오류가 남
        i = float(p_name)
    except:
        break
    else:
        print("이름을 올바르게 입력하세요")

while True:
    try:
        #나이는 무조건 숫자
        p_age = int(input("나이를 입력하세요 :"))
    except:
        print("나이를 올바르게 입력하세요")
    else:
        break


fhand = open("classoutput.txt",'w')
fhand.write("/<%s/>"%p_name)
print("<%s>"%p_name)
fhand.write("나이 : %d 세"%p_age)
print("나이 : %d 세"%p_age)
fhand.close()
```

```python
# 클라이언트가 갑자기 연락해서 분명 계약서에는 없었던 기능을 요구한다.
# 메일 주소랑 전화번호도 같이 수집해서 아래와 같이 저장해달라 한다

# e메일 주소 : XXXXX
# 전화번호 : XX-XX-X
# 생겨먹은게 완전히 다르니 함수를 만든게 의미가 없어졌다. (함수는 일반적으로 '다형성'이 없다)

#입력된게 숫자인지 판별해 리턴하는 함수
def check_int_input(inputmsg):
#inputmsg => 화면에 띄울 명령어
#isint => 숫자면 True, 아니면 False
#value => 입력된 값
    inp = input(inputmsg)
    try:
        res = int(inp)
    except: #문자면
        return (False,inp)
    else: #숫자면
        return (True,res)


#본문은 훨 간결해졌다
while True:
    check, p_name = check_int_input("이름을 입력하세요 :")
    if not check: #숫자가 아니면 ok
        break
    print("이름을 올바르게 입력하세요")
        
while True:
    check, p_age = check_int_input("나이를 입력하세요 :")
    if check: #숫자면 ok
        break
    print("나이를 올바르게 입력하세요")

#그런데 더 써먹을수가 없다

while True:
    #e메일 전화번호는 정규식 쓰는게 맞겠지만 그냥 넘어갑시다
    p_mail = input("메일주소를 입력하세요 :")
    if('@' in p_mail):
        break
    else:
        print("메일주소를 올바르게 입력하세요")
        
while True:
    p_tel = input("전화번호를 입력하세요 (-포함) :")
    #-는 두번만 쓰여야 함. 따라서 3덩이가 아니면 재입력
    if(len(p_tel.split('-')) != 3):
        print("전화번호를 올바르게 입력하세요")
        continue
    #3덩이 모두 숫자여야 함
    for s in p_mail.split('-'):
        try:
            int(s)
        except:
            print("전화번호를 올바르게 입력하세요")
    #여기까지 살아있으면 똑바로 입력된거
    break


fhand = open("classoutput.txt",'w')
fhand.write("<%s>"%p_name)
fhand.write("나이 : %d 세"%p_age)
fhand.write("메일주소 : %s"%p_mail)
fhand.write("전화번호 : %s"%p_tel)
fhand.close()

#정규식을 쓰면 뭐 어찌저찌 될거같지만... 좋은 코드는 아닌거같다.
```

```python
#클래스를 이용하면 재사용도 되고, 본문은 말도 안되게 간결해질거 같다.
#상속까지 쓸 수 있겠다.

#asker 클래스 : 인풋을 받아 내부적으로 판단한 뒤 올바른 결과값이 나올때까지 돌리는 클래스
#나는 니가 내부적으로 뭘 하는지 관심도 없어!
#너는 그저 물어본 값이 올바른 값일 때까지 유저를 붙잡아두었다가 결과값만 뱉어주면 돼!

class Asker:
    def __init__(self,s_msg="",f_msg="",ask_msg=""):
        #메세지 초기화
        self.successmessage = s_msg
        self.failmessage = f_msg
        self.askmessage = ask_msg

    def check_value(self,value):#올바른 값인지 확인하는 메서드
        return True
        #부모 클래스에서는 그냥 무슨 value건 ok
    
    def process(self,value):#값 후처리 메서드
        return value #일단 부모 클래스에서는 그냥 리턴

    def ask(self):#사용자에게 질문해서 값을 받는 메서드
        while True: #무한루프
            s = input(self.askmessage)
            if self.check_value(s): #값을 체크해서 올바른 값이면
                self.output = s
                if (self.successmessage != ""):
                    print(self.successmessage)
                break
            else:
                if (self.failmessage != ""):
                    print(self.failmessage)
        return self.process(self.output) #값 후처리 후 내보냄

    
#여기까지 만든게 뭔고 하니...
#asker는 asker인데 검사는 무조건 true
#후처리도 없음
#...뭔가 배보다 배꼽이 더 커보인다... 싶겠지만 이제부터 시작입니다





#문자인지 알아내는 asker
class StrAsker(Asker):
    #float 바꿔보고 안되면 문자아!
    def check_value(self,value):
        try:
            i = float(value)
        except:
            return True
        else:
            return False
    #후처리는 할 필요 없겠네?

    
#정수인지 알아내는 asker
class IntAsker(Asker):
    #정수면 정수 전환이 되겠지?
    def check_value(self,value):
        try:
            i = int(value)
        except:
            return False
        else:
            return True
    #후처리로 정수로 바꿔줘야 해!
    def process(self,value):
        return int(value)

#특정 문자 포맷인지 split으로 알아내는 asker
class SplitAsker(Asker):
    #split할 문자와 split 후 덩어리 숫자를 따로 받자!
    def __init__(self,s_msg="",f_msg="",ask_msg="",split_char="",split_group_num=0,input_is_num=True):
        #메세지 초기화
        super().__init__(s_msg,f_msg,ask_msg)#Asker의 init 실행임! (super()는 상속받은 부모 클래스)
        #Splitasker에서 추가로 초기화해야 할 사항은 아래 것들
        self.splitchar = split_char
        self.splitgroupnum = split_group_num
        self.inputisnum = input_is_num

    #정수와 문자 모드가 따로 있네?
    def check_value(self,value):
        #먼저 덩어리 갯수를 확인하자
        if len(value.split(self.splitchar)) != self.splitgroupnum:
            return False
        #그 다음은 값 종류 확인!
        for s in value.split(self.splitchar):
            if self.inputisnum:#정수모드
                try:
                    i = int(s)
                except:
                    return False
                else:
                    continue
            else:#문자모드
                try:
                    i = int(s)
                except:
                    continue
                else:
                    return False
        #for문 모두 돌았다면 타입이 올바른거지?
        return True
        

#이제 메인 프로그램이 어떻게 생겼나 하니...

asker = StrAsker(f_msg = "이름을 올바르게 입력하세요",ask_msg="이름을 입력하세요 : ")
p_name = asker.ask()#난 니가 내부적으로 뭘 하는지는 관심없어! 너는 그저 묻기만 해!
asker = IntAsker(f_msg = "나이를 올바르게 입력하세요",ask_msg="나이를 입력하세요 : ")
p_age = asker.ask()
asker = SplitAsker(f_msg = "메일주소를 올바르게 입력하세요",ask_msg="메일주소를 입력하세요 : ",split_char="@",split_group_num=2,input_is_num=False)
p_mail = asker.ask()
#전화번호로 바꿀땐 다시 만들지 말고 값만 바꾸자
#==========================================
asker.failmessage="전화번호를 올바르게 입력하세요"
asker.askmessage="전화번호를 입력하세요"
asker.splitchar="-"
asker.splitgroupnum=3
asker.inputisnum = True
#==========================================
p_tel = asker.ask()


fhand = open("classoutput.txt",'w')
fhand.write("<%s>"%p_name)
fhand.write("나이 : %d 세"%p_age)
fhand.write("메일주소 : %s"%p_mail)
fhand.write("전화번호 : %s"%p_tel)
fhand.close()
```
