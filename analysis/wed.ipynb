# Data Analysis -> 5-year climate Analysis

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
import seaborn as sns; sns.set()

"modules are successfully imported"
```

```python
# csv 업로드
yearsBase, meanBase = np.loadtxt('data/5-year-mean-1951-1980.csv', delimiter=',', usecols=(0, 1), unpack=True)
years, mean = np.loadtxt('data/5-year-mean-1882-2014.csv', delimiter=',', usecols=(0, 1), unpack=True)
```

```python
# 기본 데이터 플로팅
plt.scatter(yearsBase, meanBase)
plt.title('scatter plot of mean temp difference vs year')
plt.xlabel('years', fontsize=12)
plt.ylabel('mean temp difference', fontsize=12)
plt.show()
```

## NumPy를 통한 회귀분석 실행

```python
# polyfit을 통해 회귀분석 실행
m,b = np.polyfit(yearsBase, meanBase, 1)

# 1차함수로 리턴되게 됨(m=기울기, b=)
# 해당 값을 이용해 함수 선언
def f(x):
    return m*x + b

# 분산플롯을 실행함
plt.scatter(yearsBase, meanBase)
plt.plot(yearsBase, f(yearsBase)) #다만 아까 위에서 만든 직선도 그림
plt.title('scatter plot of mean temp difference vs year')
plt.xlabel('years', fontsize=12)
plt.ylabel('mean temp difference', fontsize=12)
plt.show()

# Prints text to the screen showing the computed values of m and b
print(' y = {0} * x + {1}'.format(m, b))
plt.show()
```

```python
# 이번에는 선형회귀분석
model = LinearRegression(fit_intercept=True)

# 모델을 만들고 그를 이용해 예측값 생성
model.fit(yearsBase[:, np.newaxis], meanBase)#모든 x축 년도에 대해서 실행
mean_predicted = model.predict(yearsBase[:, np.newaxis])#만들어진 모델에 대해 예측 실행

# 아까처럼 또 플롯
plt.scatter(yearsBase, meanBase)
plt.plot(yearsBase, mean_predicted)
plt.title('scatter plot of mean temp difference vs year')
plt.xlabel('years', fontsize=12)
plt.ylabel('mean temp difference', fontsize=12)
plt.show()

print(' y = {0} * x + {1}'.format(model.coef_[0], model.intercept_))
```

## Seaborn사용한 선형회귀

```python
# 일단은 플롯을 함
plt.scatter(years, mean)
plt.title('scatter plot of mean temp difference vs year')
plt.xlabel('years', fontsize=12)
plt.ylabel('mean temp difference', fontsize=12)
sns.regplot(yearsBase, meanBase,truncate=False)
plt.show()
```
