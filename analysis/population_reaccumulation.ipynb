```python

#원모어타임
stt = timer()
DATA_FILE_PREFIX = "생활인구/"
for year in range(2018,2022): #년도는 18,19,20,21년
    for month in range(1,13): #달은 12월까지
        st = timer()
        file_to_read = "%s%04d%02d.csv"%(DATA_FILE_PREFIX,year,month)
        try:
            data = pd.read_csv(file_to_read, index_col=False,encoding='euc-kr') #파일명
            print(file_to_read+"를 작업중")
            sumdata = data.groupby(by=['행정동코드'], as_index=False).mean() #행정동코드 기준으로 묶어서 부분합
            sumdata["시간대구분"]=year*100+month #시간대는 yyyymm으로 설정
            sumdata["기준일ID"]=year*100+month #기준일ID는 부분합하면 의미없음. 날려
            #파일 저장. 앞쪽 거추장한거 다 떼어버리고 파일명 설정
            #공공데이터는 UTF표준 안지키고 한국 독자규격씀
            sumdata.to_csv("생활인구/%04d%02d_acc.csv"%(year,month),encoding="euc-kr") 
            elapsed = timer() - st       
            print("변환 완료 : %s => 생활인구/%04d%02d_acc.csv (%d초 소모)"%(file_to_read,year,month,elapsed))
        except:
            print("변환 실패 : %s"%file_to_read)
            continue
print("총 소모시간 : %d"%(timer()-stt))
```
