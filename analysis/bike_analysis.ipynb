```python
import pandas as pd

df = pd.read_csv("data/bikedata.csv")
df
```

```python
import matplotlib.pyplot as plot

cause = df.drop("rentals",axis=1)
cause = cause.drop("date",axis=1)
cause
```

```python

# This ensures plots are displayed inline in the Jupyter notebook
%matplotlib inline

for col in cause.columns[1:]:
    fig = plot.figure(figsize=(9, 6))
    ax = fig.gca()
    feature = df[col]
    label = df['rentals']
    correlation = feature.corr(label)
    plot.scatter(x=feature, y=label)
    plot.xlabel(col)
    plot.ylabel('rentals')
    ax.set_title('rentals vs %s'%col)
    fig.show()
    
```

```python
df['day'] = pd.DatetimeIndex(df['date']).strftime('%A')
df = df.join(df.day.str.get_dummies())
df.head(5)
```

```python


fig = plot.figure(figsize=(9, 6))
ax = fig.gca()
feature = df['day']
label = df['rentals']
plot.scatter(x=feature, y=label)
plot.xlabel('day')
plot.ylabel('rentals')
ax.set_title('rentals vs day')
fig.show()
    
```

```python

temp = df[['rentals','day']].groupby('day').mean()
temp
```

```python
df['month'] = pd.DatetimeIndex(df['date']).strftime('%m')
temp = df[['rentals','month']].groupby('month').mean()

barlist = temp.plot.bar()

plot.title('Rentals per month')
plot.xlabel('month')
plot.ylabel('Number of Rentals')
plot.show()
```
